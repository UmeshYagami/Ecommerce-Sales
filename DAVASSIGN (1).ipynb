```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is named df
df['Price Each'] = df['Price Each'].astype(float)
price_groups = df['Price Each'].apply(lambda x: 'Greater than 100' if x > 100 else 'Less than or equal to 100')
price_groups_counts = price_groups.value_counts()

# Create a bar chart
plt.bar(price_groups_counts.index, price_groups_counts.values)
plt.xlabel('Price Range')
plt.ylabel('Number of Products')
plt.title('Number of Products by Price Range')
plt.show()
```

```python
df['Price Each'] = pd.to_numeric(df['Price Each'], errors='coerce')

# Define price ranges
price_ranges = [(0, 50), (50, 100), (100, 150), (150, 200), (200, float('inf'))]

# Create a new column 'Price Range' based on the price ranges
df['Price Range'] = pd.cut(df['Price Each'], bins=[r[0] for r in price_ranges] + [float('inf')], labels=[f'${r[0]}-${r[1]}' for r in price_ranges])

# Group by 'Product' and 'Price Range', then count the number of unique products
product_count = df.groupby('Price Range')['Product'].nunique()

# Plot the data
product_count.plot(kind='bar', figsize=(12, 6))
plt.title('Number of Products in Different Price Ranges')
plt.xlabel('Price Range')
plt.ylabel('Number of Products')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
df['Price Each'] = pd.to_numeric(df['Price Each'], errors='coerce')

# Define price ranges
price_ranges = [(0, 50), (50, 100), (100, 150), (150, 200), (200, float('inf'))]

# Create a new column 'Price Range' based on the price ranges
df['Price Range'] = pd.cut(df['Price Each'], bins=[r[0] for r in price_ranges] + [float('inf')], labels=[f'${r[0]}-${r[1]}' for r in price_ranges])

# Group by 'Price Range' and count the number of unique products
product_count = df.groupby('Price Range')['Product'].nunique()

# Create a pie chart
plt.figure(figsize=(10, 8))
plt.pie(product_count, labels=product_count.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.title('Distribution of Products in Different Price Ranges')
plt.show()
```

```python
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'], errors='coerce')

# Group by product and sum the quantity ordered
product_sales = df.groupby('Product')['Quantity Ordered'].sum().reset_index()

# Sort the products by the total quantity ordered in descending order
product_sales = product_sales.sort_values(by='Quantity Ordered', ascending=False)

# Create a bar chart
plt.figure(figsize=(12, 8))
plt.bar(product_sales['Product'], product_sales['Quantity Ordered'], color='skyblue')
plt.xlabel('Product')
plt.ylabel('Total Quantity Ordered')
plt.title('Most Purchased Products')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
